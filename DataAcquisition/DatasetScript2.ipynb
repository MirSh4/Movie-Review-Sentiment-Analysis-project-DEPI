```python
review_dict = {
    'review_id': [],
    'review': [],
    'movie_title': [],
    'tconst':[],
    'rating': []

}
```

```python
def save_progress():
    global review_dict 
    df = pd.DataFrame(review_dict)
    review_dict = {
    'review_id': [],
    'review': [],
    'movie_title': [],
    'tconst':[],
    'rating': []
}
    df.to_csv('scraped_reviews_partial.csv',mode='a',header=False, index=False)
    print("Progress saved to scraped_reviews_partial.csv!")

len(tconst)
```

```python
for a in range(tconst.index(id),len(tconst)) :
    url=fr'https://www.imdb.com/title/{tconst[a]}/reviews/?ref_=tt_urv'
    
    for attempt in range(5):
        try:
            response = requests.get(url, headers=headers)
            soup = BeautifulSoup(response.content, 'html.parser')
            movie_title = soup.title.text.split('-')[0].strip()
            
            reviews = soup.find_all('div', class_='ipc-html-content-inner-div', role='presentation', limit=30)
            if not reviews:
                print(f"No reviews found for movie: {movie_title}.")
                break
            else:
                for review in reviews:
                    review_text = review.get_text(strip=True)
                
                    review_dict['tconst'].append(tconst[a])
                    review_dict['review_id'].append(review_id_)
                    review_dict['review'].append(review_text)
                    review_dict['movie_title'].append(movie_title)
                    review_dict['rating'].append(None)
                    review_id_ += 1

            print(f"Scraped 30 reviews for movie: {movie_title}")
            save_progress()  # Save progress after scraping each movie
            break  # Exit retry loop if successful
        except ConnectionError:
            print(a)
            print(f"Connection failed, retrying {attempt + 1}/5 for {movie_title}...")
            time.sleep(5)  # Wait for 5 seconds before retrying
    else:
        print(f"Max retries exceeded for movie: {a}. Moving to next movie...")


```
